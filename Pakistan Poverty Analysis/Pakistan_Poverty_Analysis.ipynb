# **Pakistan Poverty Analysis From 2000 to 2023**

```python
import pandas as pd
import seaborn as sns
import numpy as np 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Pakistan_Poverty_Dataset_2000_2023.csv")
```

```python
fi,ax = plt.subplots(1,2, figsize=(16,5))
                                   
#plot1
ax[0].plot(df['Year'],df['Unemployment Rate (%)'],color='blue')
ax[0].set_xlabel('Year')
ax[0].set_ylabel('Unemployment Rate (%)')
ax[0].set_title('Unemployement over Time')

#plot2
ax[1].plot(df['Year'],df['Poverty Headcount Ratio (%)'],color='red')
ax[1].set_xlabel('Year')
ax[1].set_ylabel('Poverty Headcount Ratio (%)')
ax[1].set_title('Poverty Over Time')

plt.show()
```

## **Insight:**
 * Unemployment rate was highest in 2020 due to COVID also it rose after 2010 floods while Poverty Headcount ratio was lowest in 2018 and rose after 2020.

```python
fi,ax = plt.subplots(1,2, figsize=(16,5))
                                   
#plot1
ax[0].plot(df['Year'],df['GDP Growth Rate (%)'],color='green')
ax[0].set_xlabel('Year')
ax[0].set_ylabel('GDP Growth Rate (%)')
ax[0].set_title('GDP Growth Over Time')

#plot2
ax[1].plot(df['Year'],df['Government Social Spending (% of GDP)'],color='orange')
ax[1].set_xlabel('Year')
ax[1].set_ylabel('Government Social Spending (% of GDP)')
ax[1].set_title('GDP Social Spending Over Time')

plt.show()
```

## **Insight:**
 * GDP Growth was highest in 2005 while lowest in 2020 while Social Spending of GDP percentage is almost showing linear behaviour.

```python
fi,ax = plt.subplots(1,2, figsize=(16,5))
                                   
#plot1
ax[0].plot(df['Year'],df['Inflation Rate (%)'],color='black')
ax[0].set_xlabel('Year')
ax[0].set_ylabel('Inflation Rate (%)')
ax[0].set_title('Inflation Over Time')

#plot2
ax[1].plot(df['Year'],df['Agriculture Growth Rate (%)'],color='purple')
ax[1].set_xlabel('Year')
ax[1].set_ylabel('Agriculture Growth Rate (%)')
ax[1].set_title('Agricultural Growth Over Time')

plt.show()
```

## **Insight:**
 * Inflation rate was highest around 2007 t0 2008 while Agricultural growth was lowest in 2020 due to COVID and then rose significantly.

```python
plt.figure(figsize=(8,5))
plt.plot(df['Year'],df['Unemployment Rate (%)'],label='Unemployment Rate (%)')
plt.plot(df['Year'],df['Poverty Headcount Ratio (%)'],label='Poverty Headcount Ratio (%)')
plt.plot(df['Year'],df['GDP Growth Rate (%)'],label='GDP Growth Rate (%)')
plt.plot(df['Year'],df['Government Social Spending (% of GDP)'],label='Government Social Spending (% of GDP)')
plt.plot(df['Year'],df['Inflation Rate (%)'],label='Inflation Rate (%)')
plt.plot(df['Year'],df['Agriculture Growth Rate (%)'],label='Agriculture Growth Rate (%)')
plt.xlabel('Year')
plt.ylabel('Percentage')
plt.title("Comparison")
plt.legend()
plt.tight_layout()
plt.show()
        
```

## **Insight:**
 * Comparison showing  highest Poverty Headcount Ratio after 2020 followed by inflation rate

```python
plt.hist(df['Unemployment Rate (%)'],color='red',bins=12,edgecolor='black')
plt.title("Distribution of Unemployment")
plt.xlabel('Unemployment Rate')
plt.tight_layout()
plt.show()
```

```python
before_COVID = df[df['Year']<2020]
print('Unemployment rate before Covid :',before_COVID)
after_COVID  = df[df['Year']>=2020]
print('Unemployment rata after Covid :',after_COVID)
```

```python
plt.title("Unemployment Before & After COVID")
labels=['Before COVID','After COVID']
values=[
    before_COVID['Unemployment Rate (%)'].values[0],
    after_COVID['Unemployment Rate (%)'].values[0]
]
plt.bar(labels,values,color='yellow',edgecolor='black')
plt.show()
```

## **Insight:**
 * Unemployment After COVID rose significantly
